## Step 1 — Open NCBI

Open a new browser tab and go to:

👉 **https://www.ncbi.nlm.nih.gov**

This is the **National Center for Biotechnology Information** — the main public archive for biological sequences in the world.

**Take a look around.** What do you notice on the homepage? You'll see a search bar at the top — that's where most journeys begin.


## Step 2 — Search for a gene

We'll start with a famous one: **BRCA1** (the breast cancer gene).

1. In the search bar, change the dropdown from "All Databases" to **Gene**
2. Type **BRCA1** and press Enter
3. Look at the results — you'll see entries from many species
4. Click the entry for **Homo sapiens (human)** — it should be at the top

**You're now looking at the human BRCA1 gene record.** Take a minute to scroll through it.


## Step 3 — Read the gene summary

On the BRCA1 gene page, find these things and write them down in your notes:

1. What chromosome is BRCA1 on?
2. How many exons does it have?
3. What is its "Gene ID"?
4. What does the **Summary** section say about the gene's function?



## Step 4 — Download the FASTA

Now we'll get the actual sequence data.

1. On the BRCA1 page, scroll down to **NCBI Reference Sequences (RefSeq)**
2. Find the **mRNA and Protein(s)** section
3. Click on the mRNA accession that starts with **NM_** (e.g., **NM_007294**)
4. On the new page, look for the **FASTA** link near the top — click it
5. You'll see a page with the sequence. Use **File → Save As** in your browser to save it
   - Save it as `BRCA1.fasta` somewhere you can find it

**Congratulations — you just downloaded real biological data from the world's largest database.** 🎉


## Step 5 — Open the FASTA file in a text editor

This is the most important step in this whole practical.

1. Find the `BRCA1.fasta` file on your computer
2. Right-click it → **Open with → Notepad** (Windows) or **TextEdit** (Mac)
3. Look at it carefully

**What do you see?**

- The first line starts with `>` — that's the **header**. It contains the accession number, species, and description.
- Every other line is the actual DNA sequence — just letters A, T, G, C.

👉 **This is what bioinformatics data looks like in the wild.** No magic, no special software needed — just a text file you can open in Notepad.


## Step 6 — Compare GenBank format

FASTA is the simplest format. Let's see a richer one — **GenBank format**.

1. Go back to the NM_007294 page
2. Look near the top — you'll see format options (FASTA, GenBank, GenBank (full), etc.)
3. Click **GenBank**

**Compare what you see with the FASTA you saved.**

Notice:
- GenBank has **metadata** — author, journal, date, organism
- It has a **FEATURES** section that lists every gene part — exons, introns, CDS, etc.
- The sequence itself is at the bottom, in the **ORIGIN** section

Same data — but GenBank tells you a *story* about the sequence, while FASTA just gives you the letters.


## Step 7 — Find the protein

Every gene that codes for a protein has a corresponding protein record.

1. Go back to the BRCA1 gene page (you can use your browser's back button)
2. In the same RefSeq section, find a protein accession that starts with **NP_** (e.g., **NP_009225**)
3. Click it
4. You're now on the **protein record**. Click **FASTA** at the top

**Compare the protein FASTA to the DNA FASTA.**

What's different?
- The DNA had only 4 letters (A, T, G, C)
- The protein has **20 different letters** — one for each amino acid
- The protein is much shorter than the DNA


## Step 8 — Visit UniProt

NCBI is great for sequences. **UniProt** is the gold standard for protein information.

1. Open a new tab: **https://www.uniprot.org**
2. In the search bar, type **BRCA1** and press Enter
3. Find the **reviewed (Swiss-Prot)** entry for **Homo sapiens**
4. Click it (the accession will be **P38398**)

Look around this page — it has *much more* than NCBI on a protein:

- Function and biological role
- Where in the cell it's located
- Disease associations
- Known mutations
- 3D structure links
- Related publications

**This is the encyclopedia of proteins.** When you need to know anything about a protein — start here.


## Step 9 — Your turn: pick your own gene

Now do the same thing with a gene that interests **you**.

Choose one based on your background:


### Tasks for your gene

1. Find it on NCBI Gene
2. Note its **chromosome** and **Gene ID**
3. Download its **mRNA FASTA** file
4. Open the FASTA in a text editor
5. Find its protein on UniProt
6. **Note its UniProt accession** (e.g., P12345)
7. Read one sentence from its Function section

Keep all this info — you'll use it in **Part 2** when we use Python to query the same data.
