## Part 3 — Counting Nucleotides

One of the most basic analyses you can do on a DNA sequence is simply count each type of base. This tells you the composition of the sequence.

Python strings have a handy method called `.count()` that counts how many times a letter (or substring) appears.


````python
dna = "ATGGATTTATCTGCTCTTCGCGTTGAAGAAGTACAAAATGTCATTAATGCTATGCAGAAAATCTTAGAGTGTCCCATCTGTCTGGAGTTGATCAAGGAACCTGTCTCCACAAAGTGTGACCACATATTTTGCAAATTTTGC"

a_count = dna.count("A")
t_count = dna.count("T")
g_count = dna.count("G")
c_count = dna.count("C")

print("A:", a_count)
print("T:", t_count)
print("G:", g_count)
print("C:", c_count)
print("Total:", a_count + t_count + g_count + c_count)
````

### Quick check

Does the total match the length of the sequence? If yes, we counted every base. If not, there's something unexpected in the sequence (like an `N` for "unknown" or a formatting issue). This is a small but important sanity check.

## Part 4 — GC Content

**GC content** is the percentage of bases in a DNA sequence that are either G or C. It's a surprisingly useful number:

- GC content varies between organisms (some bacteria have very high GC, others very low)
- Regions of a genome with unusual GC content often mark genes, regulatory regions, or signs of evolution
- In the lab, GC content affects how DNA behaves (PCR, melting temperature, etc.)

Let's calculate it.


````python
dna = "ATGGATTTATCTGCTCTTCGCGTTGAAGAAGTACAAAATGTCATTAATGCTATGCAGAAAATCTTAGAGTGTCCCATCTGTCTGGAGTTGATCAAGGAACCTGTCTCCACAAAGTGTGACCACATATTTTGCAAATTTTGC"

g_count = dna.count("G")
c_count = dna.count("C")
gc_count = g_count + c_count

gc_percent = (gc_count / len(dna)) * 100

print("GC content: {:.2f}%".format(gc_percent))
````

### Your Turn ✏️

Write a small function that takes any DNA sequence and returns its GC content. Functions are just reusable recipes.

**Hint:** The code is already written for you below — just run it, then try it on a different sequence.


````python
def gc_content(sequence):
    """Return the GC content of a DNA sequence as a percentage."""
    g = sequence.count("G")
    c = sequence.count("C")
    return (g + c) / len(sequence) * 100

# Try it on a few different sequences
print("Sequence 1 GC%:", gc_content("ATGCATGC"))
print("Sequence 2 GC%:", gc_content("AAAAATTTTT"))
print("Sequence 3 GC%:", gc_content("GCGCGCGCGC"))

# Now try your own sequence
my_sequence = "ATGCGTACGTAGCTAGCTAGC"  # Change this
print("My sequence GC%:", gc_content(my_sequence))
````

## Part 5 — Finding a Motif

A **motif** is a short pattern in a sequence that has biological meaning. One of the most famous motifs in biology is the **start codon**: `ATG`.

Every gene that codes for a protein begins with `ATG` at the start of its coding region — this is the signal that says "start making protein here" to the cell's ribosome.

Let's find where `ATG` occurs in our BRCA1 sequence.


````python
dna = "ATGGATTTATCTGCTCTTCGCGTTGAAGAAGTACAAAATGTCATTAATGCTATGCAGAAAATCTTAGAGTGTCCCATCTGTCTGGAGTTGATCAAGGAACCTGTCTCCACAAAGTGTGACCACATATTTTGCAAATTTTGC"
motif = "ATG"

# Count how many times the motif appears
count = dna.count(motif)
print("The motif", motif, "appears", count, "times in the sequence.")

# Find the position of each occurrence
print("\nPositions (0-based index):")
start = 0
while True:
    position = dna.find(motif, start)
    if position == -1:
        break
    print("  Position", position)
    start = position + 1
````

### Why this matters

In a real genome, tools like BLAST and gene prediction algorithms do exactly this — but on millions of bases, looking for many different motifs at once. You just did the building block of gene finding.

## Part 6 — Working with FASTA Files

In the real world, biological sequences are stored in files. The most common format is called **FASTA**. A FASTA file looks like this:

```
>gene_name Some description
ATGGATTTATCTGCTCTTCGCGTTGAAGAAGTACAA
AATGTCATTAATGCTATGCAGAAAATCTTAGAGTGT
CCCATCTGTCTGGAGTTGATCAAGGAACCTGTCTCC
```

- The first line starts with `>` and gives the sequence a name and description (called the **header**).
- The following lines contain the actual sequence, often wrapped at 60 or 80 characters per line.

Let's create a small FASTA file and read it in.


````python
# Write a small FASTA file (pretending we downloaded it)
fasta_content = """>BRCA1_partial Human breast cancer 1 gene (partial)
ATGGATTTATCTGCTCTTCGCGTTGAAGAAGTACAAAATGTCATTAATGCTATGCAGAAA
ATCTTAGAGTGTCCCATCTGTCTGGAGTTGATCAAGGAACCTGTCTCCACAAAGTGTGAC
CACATATTTTGCAAATTTTGC"""

with open("brca1.fasta", "w") as f:
    f.write(fasta_content)

print("Saved brca1.fasta")
````

````python
# Now read it back
with open("brca1.fasta", "r") as f:
    lines = f.readlines()

# The first line is the header
header = lines[0].strip()
print("Header:", header)

# All other lines are the sequence — join them into one string
sequence = "".join(line.strip() for line in lines[1:])
print("Sequence length:", len(sequence), "bp")
print("First 60 bases:", sequence[:60])
````

### Using Biopython (the shortcut)

In practice, bioinformaticians don't parse FASTA files by hand — they use a library called **Biopython** that handles all the common formats. Let's install it and try.


````python
# Install Biopython (Colab has many packages preinstalled, but not this one)
!pip install biopython --quiet
````

````python
from Bio import SeqIO

# SeqIO can read FASTA and many other formats with one function
for record in SeqIO.parse("brca1.fasta", "fasta"):
    print("ID:", record.id)
    print("Description:", record.description)
    print("Length:", len(record.seq), "bp")
    print("GC content:", gc_content(str(record.seq)), "%")
    print("First 60 bases:", record.seq[:60])
````
